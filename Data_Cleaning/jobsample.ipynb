# Missing Data Imputation

For the following part im going to be filling NaNs with Unknown where it makes sense

```python
df["Start Date"]=df["Start Date"].fillna("Unknown") 
df["Start Date"].isnull().sum()
```

```python
df["End Date"]=df["End Date"].fillna("Unknown")
df["End Date"].isnull().sum()
```

```python
df["Campaign categories"]=df["Campaign categories"].bfill()
df["Campaign categories"].isnull().sum()
```

```python
df["Primary Affiliation: Account Name"]=df["Primary Affiliation: Account Name"].fillna("Unknown")
df["Primary Affiliation: Account Name"].isnull().sum()
```

```python
###I’m going to use the email domain’s ccTLD to infer the country for records with null values.

ccTLD = {"pe" : "Perú", "co" : "Colombia" , "mx" : "México" , "ar" : "Argentina" , "cl" : "Chile" , "br" : "Brasil", "uk" : "United Kingdom" , "gt" : "Guatemala", "pa" : "Panamá"}

for x in df["Email"].items():
    tld_parts=x[1].rsplit(".",1) #list with index and mail splitted eg(direccion@fec.org) (pa) this way i can use whats next to the point to infer the country 
    tld=tld_parts[1]
    idx = x[0]
    country=ccTLD.get(tld)
    df.at[idx, "Country"]
    if country is not None and df["Country"].isna().at[idx]: 
      df.at[idx, "Country"]=country 

df.head(30)     
```

```python
### Because the dataset includes four country fields, I assume they refer to the same country per record.
df["Country"]=(
    df["Country"]
    .fillna(df["Billing Country"])
    .fillna(df["Country Presence"])
    .fillna(df["Country.1"])
)
df["Billing Country"]=(
    df["Billing Country"]
    .fillna(df["Country"])
    .fillna(df["Country Presence"])
    .fillna(df["Country.1"])
)
df["Country Presence"]=(
    df["Country Presence"]
    .fillna(df["Country"])
    .fillna(df["Billing Country"])
    .fillna(df["Country.1"])
)

df["Country"]=df["Country"].fillna("Unknown")
df["Billing Country"]=df["Billing Country"].fillna("Unknown")
df["Country Presence"]=df["Country Presence"].fillna("Unknown")
df.pop("Country.1")
df.head(30)
```

```python
#Im standarizing Country names, i found that there are cases where accents are not used, also for USA im using the most repeated way used
df["Country"]=(
    df["Country"]
   .replace({"Peru":"Perú","Mexico":"México", "Panama":"Panamá","USA" : "Estados Unidos", "United States" : "Estados Unidos"})
)
df["Billing Country"]=(
    df["Billing Country"]
   .replace({"Peru":"Perú","Mexico":"México", "Panama":"Panamá" ,"USA" : "Estados Unidos" , "United States" : "Estados Unidos"})
)
df["Country Presence"]=(
    df["Country Presence"]
   .replace({"Peru":"Perú","Mexico":"México", "Panama":"Panamá","USA" : "Estados Unidos", "United States" : "Estados Unidos"})
)
```

```python
df["Origin"]=df["Origin"].fillna("Unknown")
df["Social Cause"]=df["Social Cause"].fillna("Unknown")
df.pop("#NAME?") #We delete this column with no data
df
```

# API Script

```python
#Im using the world bank API to obtain information on population, internet usage, and income levels to enrich the dataset, these functions will help me to create a for cycle to access information needed
import requests
base_url = "https://api.worldbank.org/v2/"
def get_country_pop(country):
    url = f"{base_url}/country/{country}/indicator/SP.POP.TOTL?format=json&mrnev=1"
    response = requests.get(url)
    if response.status_code == 200:
        country_data = response.json()
        return country_data

def get_country_internet(country):
    url = f"{base_url}/country/{country}/indicator/IT.NET.USER.ZS?format=json&mrnev=1"
    response = requests.get(url)
    if response.status_code == 200:
        country_data = response.json()
        return country_data

def get_country_income(country):
    url = f"{base_url}/country/{country}?format=json&mrnev=1"
    response = requests.get(url)
    if response.status_code == 200:
        country_data = response.json()
        return country_data
```

```python
pop_map={} 
for x in df["Country_ISO"].unique():
    resp=get_country_pop(x)
    if resp and len(resp) > 1 and resp[1]:
        pop_map[x] = resp[1][0]["value"]

df["Population"] = df["Country_ISO"].map(pop_map)
df["Population"]=df["Population"].fillna("Unknown")
```

```python
int_map={}
for x in df["Country_ISO"].unique():
    resp=get_country_internet(x)
    if resp and len(resp) > 1 and resp[1]:
        int_map[x] = resp[1][0]["value"]

df["Internet"] = df["Country_ISO"].map(int_map)
df["Internet"]=df["Internet"].fillna("Unknown")
```

```python
income_map={}
for x in df["Country_ISO"].unique():
    resp=get_country_income(x)
    if resp and len(resp) > 1 and resp[1]:
        income_map[x] = resp[1][0]["incomeLevel"]["value"]

df["Income Level"] = df["Country_ISO"].map(income_map)
df["Income Level"]=df["Income Level"].fillna("Unknown")
```

```python
df.to_csv("enriched_salesforce_report.csv")
```
